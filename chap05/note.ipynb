```python
def sum(n): 
    hab = 0
    for i in range(1, n+1): 
        hab = hab + i 
    return hab 
a = sum(50) 
b = sum(1000) 
print(a)
print(b)
```

```python
import math 

def cir_area(r):
    return math.pi * r * r

def cir_cirm(r):
    return 2 * math.pi * r


r = 3.5
area = cir_area(r)
circumference = cir_cirm(r)

print("반지름:", r, "cm")
print("원의 면적:", round(area, 1), "cm^2")
print("원의 둘레:", round(circumference, 1), "cm")
```

```python
def func1() :
    v = 10 
    print("func1()의v = %d" % v)
def func2() :
    global v 
    v = 30
    print("func2()의변경된전역v = %d" % v)
    
v = 20 

func1() 
func2() 
print(v) 
```

```python
def total(*numbers):
    sum = 0
    for n in numbers: 
        sum += n
    return sum
print (total(1,2,3))
print (total(1,2,3,4,5)) 
```
