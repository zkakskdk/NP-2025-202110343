1. 두개의매개변수n, m을전달받아m x n개의* 상자를출력하는프로그램을함수로작성
예: 2, 4 → **** 
           ****

```python
def print_box(m, n):
    for _ in range(m):
        print("*" * n) 

print_box(2, 4) 
```

2. 하나의 숫자를 전달받아 숫자의 자리합을 구하는 함수를 작성
예: 123 → 1+2+3 = 6

```python
def digit_sum(n):
    hab = 0
    for ch in str(n):   
        hab += int(ch)  
    return hab

print(digit_sum(123)) 
print(digit_sum(9876))  
```

 3. 두개의 문자열이 서로 다른 처음위치를 반환하는 함수를 작성. 두개의 문자열이 같으면 -1을 반환

```python
def first_diff(str1, str2):
    length = min(len(str1), len(str2)) 
    
    for i in range(length):
        if str1[i] != str2[i]:
            return i   

    if len(str1) != len(str2):  
        return length  
    
    return -1  

print(first_diff("apple", "apricot"))  
print(first_diff("test", "test"))     
print(first_diff("abc", "abcd"))      
```

4. 숫자를 전달받아 그 수의 약수를 리스트로 반환하는 함수를 작성

```python
def get_divisors(n):
    divisors = []
    for i in range(1, n+1):
        if n % i == 0:  
            divisors.append(i)
    return divisors

print(get_divisors(12))  
print(get_divisors(17))  
```

5. 문자열과 하나의 문자를 전달받아 문자열에서 문자의 위치를 모두 찾아 리스트로 반환하는 함수를 작성

```python
def find_char_positions(s, ch):
    positions = []
    for i, c in enumerate(s):  
        if c == ch:
            positions.append(i)
    return positions

print(find_char_positions("hello", "l")) 
print(find_char_positions("banana", "a"))  
print(find_char_positions("apple", "z"))
```

6. 재귀함수를 이용하여 1부터 100까지의 합을 계산하는 프로그램

```python
def recursive_sum(n):
    if n == 1:       
        return 1
    else:
        return n + recursive_sum(n-1) 

result = recursive_sum(100)
print("1부터 100까지의 합:", result)
```
