 동물(Animal)을 울음을추상메소드로하는추상클래스로정의하고,
 • 개(Dog)와 고양이(Cat)를 구체적인 클래스로 정의한 후,
 • 각 동물의 울음소리를 출력하는 프로그램을 작성하시오.
 • 각 동물객체는 자신의 이름을 속성으로 가지며,
 • 생성자에서 이름을 초기화합니다.
 • 추가적으로 Animal 클래스는 현재 생성된 동물객체의 수를 클래스 변수로 가지고,
 • 정적 메소드로 현재 동물객체의 수를 반환하는 메소드를 구현
 • 객체가 생성(__init__)될 때 객체의 수를 증가하고 삭제(__del__(self) ) 될 때 객체의 수를 감소

 테스트
 • 개나 고양이객체들을 리스트로 생성하고 개나 Animal 변수로 받아 동적바인딩이 되는지 확인하시오.
 • 모든 객체들이 울게하는 함수를 작성하여 테스트하시오. 

```python
from abc import ABC, abstractmethod

# 추상 클래스 정의
class Animal(ABC):
    # 클래스 변수 (생성된 동물 수)
    _count = 0

    def __init__(self, name):
        self.name = name
        Animal._count += 1   # 객체 생성 시 증가

    def __del__(self):
        Animal._count -= 1   # 객체 삭제 시 감소

    @abstractmethod
    def cry(self):
        """추상 메소드: 동물의 울음소리"""
        pass

    @staticmethod
    def get_count():
        """현재 생성된 동물 객체 수 반환"""
        return Animal._count


# 구체 클래스: Dog
class Dog(Animal):
    def cry(self):
        print(f"{self.name}: 멍멍!")


# 구체 클래스: Cat
class Cat(Animal):
    def cry(self):
        print(f"{self.name}: 야옹~")


# 모든 동물을 울게 하는 함수
def make_all_animals_cry(animals):
    for a in animals:
        a.cry()   # 동적 바인딩으로 올바른 cry() 호출됨


# 테스트
if __name__ == "__main__":
    animals = [
        Dog("바둑이"),
        Cat("나비"),
        Dog("초코"),
        Cat("하양이")
    ]

    print("현재 동물 객체 수:", Animal.get_count())

    # 동적 바인딩 테스트 (Animal 타입 변수에 담기)
    a: Animal = animals[0]
    a.cry()   # Dog의 cry()가 호출됨

    # 모든 동물이 울게 하기
    make_all_animals_cry(animals)

    # 객체 삭제 후 카운트 확인
    del animals[0]
    del animals[1]
    print("객체 일부 삭제 후 동물 객체 수:", Animal.get_count())
```
