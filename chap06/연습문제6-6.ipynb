[출력 내용]
이름: 홍길동, 학번: 20150001, 학과: 로봇공학과
3학년입니다

```python
class Student:
    def __init__(self, name, id, dept):
        self.name = name
        self.id = id
        self.dept = dept

    def grade(self):
        year = int(str(self.id)[:4])
        if year < 2018:
            print("3학년입니다")
        else:
            print("3학년이 아닙니다")


man = Student(name="김지성", id=20150001, dept="컴퓨터공학과")
print("이름: {}, 학번: {}, 학과: {}".format(man.name, man.id, man.dept))
man.grade()
```
