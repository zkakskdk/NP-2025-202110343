```python
class Person:
    count = 1000  

    def __init__(self, name, age): 
        self.name = name 
        self.age = age

        Person.count += 1
        self.__id = Person.count  

    def getId(self): 
        print(self.__id) 

    def getAge(self): 
        print(self.age)

    def getName(self):
        print(self.name)


class Employee(Person):
    def __init__(self, name, age, employeeID):
        super().__init__(name, age)
        self.employeeID = employeeID

    def getEmployeeID(self):
        print(self.employeeID)


p1 = Person("홍길동", 20)
p2 = Person("이순신", 30)
e1 = Employee("동양", 65, 2019)

p1.getName(); p1.getId()
p2.getName(); p2.getId()
e1.getName(); e1.getId(); e1.getEmployeeID()
```
